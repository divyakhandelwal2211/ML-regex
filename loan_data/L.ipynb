```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('loan_data.csv')
df
```

```python
df.info()
```

```python
num_cols=df.select_dtypes(include='float').columns
cat_cols=df.select_dtypes(include='object').columns
```

```python
df[num_cols]
```

```python
df[cat_cols]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Here we have created the object of standard scaler
scaler = StandardScaler()

# Here we have fit and transformed the numerical columns
scaled_value = scaler.fit_transform(df[num_cols])

# Scaled Dataframe
scaled_df=pd.DataFrame(scaled_value,columns=num_cols)
scaled_df
```

```python
cat_cols
```

```python
from sklearn.preprocessing import LabelEncoder

Label_enc = LabelEncoder()

for col in cat_cols:
    df[col]=Label_enc.fit_transform(df[col])
```

```python
df[cat_cols]
```

```python
encoded_df = df[cat_cols]
# features
X = pd.concat([scaled_df,encoded_df],axis=1)
Y = df['loan_status']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```
