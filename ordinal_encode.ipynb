```python
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd
```

```python
df = pd.DataFrame({'Remark':['good','better','best','good']})

encoder = OrdinalEncoder(categories=[['good','better','best']])
df['Encoded Column'] = encoder.fit_transform(df[['Remark']])

# fit -> Learns mapping good = 0 , better=1,best=2
#Transform -> Replace the actual text with numerical value
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
import pandas as pd
df1= pd.DataFrame({'Colors':['red','green','blue','blue','green']})
```

```python
# we'll create the object of the onehotencoder class

enc = OneHotEncoder(sparse_output=False)  

encoded_data = enc.fit_transform(df1[['Colors']])
```

```python
encoded_data
```

```python
df_encoded = pd.DataFrame(encoded_data,columns = enc.get_feature_names_out(['Colors']))
print(df_encoded)
```

```python
# get dummies

df_enc = pd.get_dummies(df1,columns=['Colors'],dtype=int)
```

```python
df_enc
```

```python
# Import the min max scaler (normalization)

from sklearn.preprocessing import MinMaxScaler

df2 = pd.DataFrame({
    'Age':[10,20,30,40,50],
    'Salary':[10000,20000,30000,40000,50000]
})

# Creating the object of minmaxscaler
scaler = MinMaxScaler()

scaled_values = scaler.fit_transform(df2)
```

```python
scaled_df = pd.DataFrame(scaled_values,columns=df2.columns)
```
